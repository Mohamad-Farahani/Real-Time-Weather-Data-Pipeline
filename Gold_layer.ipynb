```python
from pyspark.sql.functions import avg, max, min, col

# Read silver table
silver_df = spark.table("silver_weather")

# Aggregate per hour (or per day if you want)
gold_df = silver_df.groupBy("hour_ts") \
    .agg(
        avg("temperature_c").alias("avg_temp"),
        max("temperature_c").alias("max_temp"),
        min("temperature_c").alias("min_temp")
    )

# Save as Delta table
gold_df.write.format("delta").mode("overwrite").saveAsTable("gold_weather")
```
