```python
from pyspark.sql.functions import col

silver_df = spark.table("bronze_weather").select(
    "city",
    col("fetch_time").cast("timestamp").alias("fetch_time_ts"),
    col("time").cast("timestamp").alias("hour_ts"),
    col("temperature_2m").alias("temperature_c"),
    (col("temperature_2m") * 9/5 + 32).alias("temperature_f"),
    "relative_humidity_2m",
    col("wind_speed_20m").alias("wind_speed_mps"),
    (col("wind_speed_20m") * 3.6).alias("wind_speed_kmh"),
    "ingestion_time"
)

silver_df.write.format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable("silver_weather")
```
